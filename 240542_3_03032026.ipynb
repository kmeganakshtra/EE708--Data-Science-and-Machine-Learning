```python

# EE708 Assignment 3
# Question 9


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram


#  K-MEANS CLUSTERING


print(" K-Means Clustering")


# Load Dataset

data1 = pd.read_csv("A3_P1.csv")
X1 = data1.values


# (a) Elbow Method

wcss = []

for k in range(1, 16):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1, 16), wcss, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Inertia)")
plt.title("Elbow Method")
plt.show()

print("Observing the elbow plot and choosing optimal k manually.")


# (b) Choosed Optimal k

optimal_k = 3

# (c) Perform Final Clustering

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
labels = kmeans_final.fit_predict(X1)
centers = kmeans_final.cluster_centers_

plt.figure()

for i in range(optimal_k):
    plt.scatter(X1[labels == i, 0], X1[labels == i, 1])

plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200)
plt.title("Final K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print("K-Means Completed\n")


# QUESTION 10: HIERARCHICAL CLUSTERING (CORRECTED VERSION)


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

print("Hierarchical Clustering")


# Load Dataset

data2 = pd.read_csv("A3_P2.csv")
X2 = data2.values


# SINGLE LINKAGE

Z_single = linkage(X2, method='single', metric='euclidean')

plt.figure()
dendrogram(Z_single)
plt.title("Single Linkage Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()


# COMPLETE LINKAGE

Z_complete = linkage(X2, method='complete', metric='euclidean')

plt.figure()
dendrogram(Z_complete)
plt.title("Complete Linkage Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()


# AVERAGE LINKAGE

Z_avg = linkage(X2, method='average', metric='euclidean')

plt.figure()
dendrogram(Z_avg)
plt.title("Average Linkage Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

print("Hierarchical Clustering ")
```
